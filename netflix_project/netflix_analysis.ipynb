## Question 1:
How many Movies and TV Shows are available in Netflix?

```python
df['type'].value_counts()
```

```python
df['type'].value_counts().plot(kind="bar")
plt.title("MOVIES VS TV SHOWS")
plt.show()
```

## Question 2:
Which countries produce the most Netflix content?

```python
df['country'].value_counts().head(10) #this gives top 10 values
```

## Question 3:
What are the most common ratings?

```python
df['rating'].value_counts()
```

```python
df['rating'].value_counts().head(10).plot(kind="bar")

plt.title("Most Common Ratings")
plt.show()
```

## Question 4:
Which year had the highest Netflix releases?

```python
df['release_year'].value_counts().head(10)
```

## Handling Missing Values

```python
df.isnull().sum()
```

```python
df=df.dropna(subset=["country"]) #Removed rows with missing country values to improve analysis accuracy.
```
